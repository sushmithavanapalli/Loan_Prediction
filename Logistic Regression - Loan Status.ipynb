```python
#Event Rates
count_LoanStatusY = len(Data_Set['Loan_Status'] == 1)
count_LoanStatusN = len(Data_Set['Loan_Status'] == 0)
percent_LoanStatY = count_LoanStatusY / (count_LoanStatusY + count_LoanStatusN) 
percent_LoanStatN = count_LoanStatusN / (count_LoanStatusN + count_LoanStatusN)
print("Percentage of Loan Status Yes:", percent_LoanStatY * 100)
print("Percentage of Loan Status No:", percent_LoanStatN * 100)
```

```python
#Logistic Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x = Data_Set.drop(['Loan_Status'], axis = 1)
y = Data_Set.Loan_Status

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#Create an instance and fit the model
logit_model = LogisticRegression()
logit_model.fit(x_train, y_train)
```

```python
#Predictions
Predictions = logit_model.predict(x_test)
```

```python
#Model Evaluation
from sklearn.metrics import classification_report

print(classification_report(y_test, Predictions))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, Predictions))
```

```python
#TP - 31074
#TN - 4377
#FP - 1720
#FN - 5066

#Accuracy
#Accuracy = (TP + TN)/Total

Accuracy = (64 + 104)/(104+64+63+64)

#Error Rate
#ErrorRate = (FP+FN)/Total

Error = (64 + 63)/(104+64+63+64)

print("Accuracy:", Accuracy*100)
print("Error:", Error*100)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, Predictions))
```

```python
#KNN Algorithm
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#Fitting KNN to the training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p =2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
